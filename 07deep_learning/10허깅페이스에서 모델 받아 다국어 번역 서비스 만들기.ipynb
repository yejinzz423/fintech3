```python
import gradio as gr
from transformers import M2M100ForConditionalGeneration, M2M100Tokenizer

# Load model & tokenizer
model_name = "facebook/m2m100_418M"
model = M2M100ForConditionalGeneration.from_pretrained(model_name)
tokenizer = M2M100Tokenizer.from_pretrained(model_name)

# ì–¸ì–´ ì½”ë“œ ë¦¬ìŠ¤íŠ¸ (Gradio Dropdownìš©)
lang_options = {
    "Korean (ko)": "ko",
    "English (en)": "en",
    "Chinese (zh)": "zh",
    "Japanese (ja)": "ja",
    "French (fr)": "fr",
    "German (de)": "de",
    "Spanish (es)": "es",
}

def translate_text(text, src_lang, tgt_lang):
    """
    M2M100 ë²ˆì—­ í•¨ìˆ˜
    """
    if not text.strip():
        return "â— ë²ˆì—­í•  ë¬¸ì¥ì„ ì…ë ¥í•˜ì„¸ìš”."

    tokenizer.src_lang = src_lang
    encoded = tokenizer(text, return_tensors="pt")

    generated_tokens = model.generate(
        **encoded,
        forced_bos_token_id=tokenizer.get_lang_id(tgt_lang)
    )

    result = tokenizer.batch_decode(generated_tokens, skip_special_tokens=True)
    return result[0]


# Gradio UI êµ¬ì„±
with gr.Blocks() as demo:
    gr.Markdown("## ğŸŒ M2M100 ê¸°ë°˜ ë‹¤êµ­ì–´ ë²ˆì—­ê¸°")

    with gr.Row():
        input_text = gr.Textbox(label="ì…ë ¥ ë¬¸ì¥", placeholder="ë²ˆì—­í•  ë¬¸ì¥ì„ ì…ë ¥í•˜ì„¸ìš”.")
    
    with gr.Row():
        src_lang = gr.Dropdown(list(lang_options.keys()), label="ì›ë³¸ ì–¸ì–´ ì„ íƒ", value="Korean (ko)")
        tgt_lang = gr.Dropdown(list(lang_options.keys()), label="ë²ˆì—­ ì–¸ì–´ ì„ íƒ", value="English (en)")

    translate_btn = gr.Button("ë²ˆì—­í•˜ê¸°")

    output_text = gr.Textbox(label="ë²ˆì—­ ê²°ê³¼")

    translate_btn.click(
        fn=lambda text, s, t: translate_text(text, lang_options[s], lang_options[t]),
        inputs=[input_text, src_lang, tgt_lang],
        outputs=output_text
    )

# ì‹¤í–‰
demo.launch()
```

```python
import torch
import numpy as np
import gradio as gr
import soundfile as sf

from transformers import M2M100ForConditionalGeneration, M2M100Tokenizer
from transformers import VitsModel, AutoTokenizer
from kokoro import KPipeline
from uroman import Uroman

# -----------------------------
# 1. ë²ˆì—­ ëª¨ë¸ & í† í¬ë‚˜ì´ì € ë¡œë“œ
# -----------------------------
MODEL_NAME = "facebook/m2m100_418M"

tokenizer = M2M100Tokenizer.from_pretrained(MODEL_NAME)
model = M2M100ForConditionalGeneration.from_pretrained(MODEL_NAME)

device = "cuda" if torch.cuda.is_available() else "cpu"
model = model.to(device)

# UIì—ì„œ ë³´ì—¬ì¤„ ì–¸ì–´ ì´ë¦„ â†” M2M100 ì½”ë“œ ë§¤í•‘
LANG_MAP = {
    "í•œêµ­ì–´ (ko)": "ko",
    "ì˜ì–´ (en)": "en",
    "ì¤‘êµ­ì–´ (zh)": "zh",
    "ì¼ë³¸ì–´ (ja)": "ja",
    "ìŠ¤í˜ì¸ì–´ (es)": "es"
}

# -----------------------------
# 2. TTS ëª¨ë¸ ë¡œë“œ
#    - ì˜ì–´: Kokoro
#    - í•œêµ­ì–´: MMS-TTS (facebook/mms-tts-kor)
# -----------------------------

# ì˜ì–´ TTS (Kokoro)
pipeline_en = KPipeline(lang_code='a')  # ì‚¬ìš©ìê°€ ì“°ë˜ ê·¸ëŒ€ë¡œ 'a'

# í•œêµ­ì–´ TTS (MMS)
ko_tts_model = VitsModel.from_pretrained("facebook/mms-tts-kor")
ko_tts_tokenizer = AutoTokenizer.from_pretrained("facebook/mms-tts-kor")
uroman = Uroman()

# -----------------------------
# 3. ë²ˆì—­ í•¨ìˆ˜
# -----------------------------
def translate(text, src_lang_label, tgt_lang_label):
    if not text.strip():
        return "í…ìŠ¤íŠ¸ë¥¼ ì…ë ¥í•˜ì„¸ìš”."

    src_lang = LANG_MAP[src_lang_label]
    tgt_lang = LANG_MAP[tgt_lang_label]

    # M2M100ì— ì‚¬ìš©í•  ì›ë¬¸ ì–¸ì–´ ì„¤ì •
    tokenizer.src_lang = src_lang

    # í† í¬ë‚˜ì´ì¦ˆ
    encoded = tokenizer(text, return_tensors="pt")
    encoded = {k: v.to(device) for k, v in encoded.items()}

    # target ì–¸ì–´ì˜ BOS í† í° ID
    forced_bos_token_id = tokenizer.get_lang_id(tgt_lang)

    # ë²ˆì—­ ìƒì„±
    generated_tokens = model.generate(
        **encoded,
        forced_bos_token_id=forced_bos_token_id,
        max_length=256
    )

    result = tokenizer.batch_decode(generated_tokens, skip_special_tokens=True)[0]
    return result

# -----------------------------
# 4. TTS í•¨ìˆ˜ë“¤
# -----------------------------
def tts_en(text: str):
    """
    ì˜ì–´ ë¬¸ì¥ì„ Kokoroë¡œ í•©ì„±í•´ì„œ (sr, audio_ndarray) ë°˜í™˜
    """
    # KokoroëŠ” ì—¬ëŸ¬ segmentë¥¼ generatorë¡œ ë‚´ë³´ëƒ„ â†’ ì „ë¶€ ì´ì–´ë¶™ì´ê¸°
    audio_segments = []
    for _, _, audio in pipeline_en(text, voice='af_heart'):
        audio_segments.append(audio)

    if not audio_segments:
        return None, None

    audio = np.concatenate(audio_segments)
    sr = 24000  # Kokoro ê¸°ë³¸ ì˜ˆì‹œ ìƒ˜í”Œë ˆì´íŠ¸
    return sr, audio

def tts_ko(text: str):
    """
    í•œêµ­ì–´ ë¬¸ì¥ì„ MMS-TTSë¡œ í•©ì„±í•´ì„œ (sr, audio_ndarray) ë°˜í™˜
    """
    roman_text = uroman.romanize_string(text)
    inputs = ko_tts_tokenizer(roman_text, return_tensors="pt")

    with torch.no_grad():
        output = ko_tts_model(**inputs).waveform

    waveform = output.squeeze(0).cpu().numpy()
    sr = ko_tts_model.config.sampling_rate  # ë³´í†µ 24000
    return sr, waveform

# -----------------------------
# 5. ë²ˆì—­ + TTS í†µí•© ë˜í¼
# -----------------------------
def translate_and_tts(text, src_lang_label, tgt_lang_label):
    translated = translate(text, src_lang_label, tgt_lang_label)
    tgt_lang = LANG_MAP[tgt_lang_label]

    # ê¸°ë³¸ê°’: ì˜¤ë””ì˜¤ ì—†ìŒ
    audio_result = None

    # ëª©í‘œ ì–¸ì–´ì— ë”°ë¼ TTS ì„ íƒ
    if tgt_lang == "en":
        sr, audio = tts_en(translated)
        if sr is not None:
            audio_result = (sr, audio)
    elif tgt_lang == "ko":
        sr, audio = tts_ko(translated)
        if sr is not None:
            audio_result = (sr, audio)

    # Gradioì—ì„œ (í…ìŠ¤íŠ¸, ì˜¤ë””ì˜¤) ë‘˜ ë‹¤ ë°˜í™˜
    return translated, audio_result

# -----------------------------
# 6. Gradio UI ì •ì˜
# -----------------------------
with gr.Blocks(title="M2M100 ë²ˆì—­ + TTS ë°ëª¨") as demo:
    gr.Markdown(
        """
        # ğŸŒ M2M100 ë²ˆì—­ + ìŒì„± í•©ì„± ë°ëª¨  
        `facebook/m2m100_418M`ìœ¼ë¡œ ë²ˆì—­í•˜ê³ ,  
        - **í•œêµ­ì–´ ê²°ê³¼**ëŠ” `facebook/mms-tts-kor`ë¡œ  
        - **ì˜ì–´ ê²°ê³¼**ëŠ” `Kokoro`ë¡œ  
        ì½ì–´ì£¼ëŠ” ì„œë¹„ìŠ¤ì…ë‹ˆë‹¤.  
        
        âœ… ì£¼ ì‚¬ìš© ì˜ˆì‹œ  
        - í•œêµ­ì–´ â†’ ì˜ì–´ (ì˜ì–´ ìŒì„± ì¬ìƒ)  
        - ì˜ì–´ â†’ í•œêµ­ì–´ (í•œêµ­ì–´ ìŒì„± ì¬ìƒ)  
        """
    )

    with gr.Row():
        with gr.Column():
            src_lang = gr.Dropdown(
                label="ì›ë¬¸ ì–¸ì–´",
                choices=list(LANG_MAP.keys()),
                value="í•œêµ­ì–´ (ko)"
            )
            
            src_text = gr.Textbox(
                label="ì…ë ¥ í…ìŠ¤íŠ¸",
                lines=5,
                placeholder="ë²ˆì—­í•  ë¬¸ì¥ì„ ì…ë ¥í•˜ì„¸ìš”."
            )

        with gr.Column():
            tgt_lang = gr.Dropdown(
                label="ë²ˆì—­ ì–¸ì–´",
                choices=list(LANG_MAP.keys()),
                value="ì˜ì–´ (en)"
            )
            
            out_text = gr.Textbox(
                label="ë²ˆì—­ ê²°ê³¼",
                lines=5
            )

            
    with gr.Row():
        out_audio = gr.Audio(
                label="ë²ˆì—­ ê²°ê³¼ ìŒì„±",
                autoplay=True
            )
    with gr.Row():
        translate_btn = gr.Button("ë²ˆì—­ + ìŒì„± ì¬ìƒ")

    # ë²„íŠ¼ í´ë¦­ ì‹œ ë²ˆì—­ + TTS í•¨ìˆ˜ í˜¸ì¶œ
    translate_btn.click(
        fn=translate_and_tts,
        inputs=[src_text, src_lang, tgt_lang],
        outputs=[out_text, out_audio]
    )

demo.launch()
```
