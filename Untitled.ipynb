```python
# ======================================================
# 0. í•„ìˆ˜ import
# ======================================================
import requests
import pandas as pd
import time
from datetime import datetime, timedelta

# ======================================================
# 1. API KEY
# ======================================================
API_KEY = "57843872c61b2852422b4870c0a618d09ae430f7"

# ======================================================
# 2. DART ê³µì‹œê²€ìƒ‰ API
# ======================================================
BASE_URL = "https://opendart.fss.or.kr/api/list.json"

# ======================================================
# 3. ë‚ ì§œ êµ¬ê°„ ìƒì„± í•¨ìˆ˜ (3ê°œì›” ë‹¨ìœ„)
# ======================================================
def make_date_ranges(start_date, end_date, months=3):
    ranges = []
    cur = start_date
    while cur < end_date:
        next_date = cur + pd.DateOffset(months=months)
        ranges.append((cur, min(next_date, end_date)))
        cur = next_date
    return ranges

# ======================================================
# 4. ì „ì²´ ì¡°íšŒ ê¸°ê°„ (ìµœê·¼ 3ë…„)
# ======================================================
end_date = datetime.today()
start_date = end_date - timedelta(days=365 * 3)

date_ranges = make_date_ranges(start_date, end_date, months=3)

# ======================================================
# 5. 3ê°œì›” ë‹¨ìœ„ ë°˜ë³µ ì¡°íšŒ
# ======================================================
results = []

for s, e in date_ranges:
    print(f"ğŸ“… ì¡°íšŒ ì¤‘: {s.date()} ~ {e.date()}")

    params = {
        "crtfc_key": API_KEY,
        "bgn_de": s.strftime("%Y%m%d"),
        "end_de": e.strftime("%Y%m%d"),
        "page_no": 1,
        "page_count": 100,
    }

    while True:
        res = requests.get(BASE_URL, params=params).json()

        if res.get("status") != "000":
            print("API ì˜¤ë¥˜:", res)
            break

        df = pd.DataFrame(res.get("list", []))

        # ğŸ”¥ ì¦ê¶Œë°œí–‰ì‹¤ì ë³´ê³ ì„œë§Œ
        df = df[df["report_nm"].str.contains("ì¦ê¶Œë°œí–‰ì‹¤ì ë³´ê³ ì„œ", na=False)]
        results.append(df)

        if params["page_no"] >= int(res.get("total_page", 1)):
            break

        params["page_no"] += 1
        time.sleep(0.3)

# ======================================================
# 6. ê²°ê³¼ í•©ì¹˜ê¸°
# ======================================================
result_df = pd.concat(results, ignore_index=True)

# ======================================================
# 7. ì»¬ëŸ¼ ì •ë¦¬ + ë§í¬ ìƒì„±
# ======================================================
result_df["rcept_dt"] = pd.to_datetime(result_df["rcept_dt"])
result_df["ë³´ê³ ì„œ_URL"] = (
    "https://dart.fss.or.kr/dsaf001/main.do?rcpNo="
    + result_df["rcept_no"]
)

result_df = result_df[
    [
        "corp_name",
        "corp_code",
        "report_nm",
        "rcept_dt",
        "rcept_no",
        "ë³´ê³ ì„œ_URL",
    ]
]

print(f"\nâœ… ìµœê·¼ 3ë…„ ì¦ê¶Œë°œí–‰ì‹¤ì ë³´ê³ ì„œ ì´ {len(result_df)}ê±´")

# ======================================================
# 8. ì—‘ì…€ ì €ì¥
# ======================================================
result_df.to_excel(
    "ìµœê·¼3ë…„_ì „ì²´_ì¦ê¶Œë°œí–‰ì‹¤ì ë³´ê³ ì„œ.xlsx",
    index=False
)

print("ğŸ“ ì—‘ì…€ ì €ì¥ ì™„ë£Œ")
result_df.head()
```
