# 전체

```python
import requests
import pandas as pd
from bs4 import BeautifulSoup as bs
from dotenv import load_dotenv
import time
import os
import re
load_dotenv("./naver_search_web/data/")
```

```python
start = 1
end = 1000
page_num = 1
total_page = 1
result2 = {}

while page_num <= total_page:
    url = f'http://openapi.seoul.go.kr:8088/615659564f776a643539557a6f4c52/xml/VwsmAdstrdNcmCnsmpW/{start}/{end}/'
    payload = dict(KEY=os.getenv("seoul"), Type="xml", pIndex=1, pSize=10)
    r = requests.get(url)
    response = bs(r.text, "xml")
    response

    total_page = int(response.select_one("list_total_count").text) // 1000+1
    tag_list = [tags.name for tags in response.select_one('row').find_all()]
    
    kor_cols = ['기준_년분기_코드', '행정동_코드', '행정동_코드_명', '월_평균_소득_금액', '소득_구간_코드',
                '지출_총금액', '식료품_지출_총금액', '의류_신발_지출_총금액', '생활용품_지출_총금액',
                '의료비_지출_총금액', '교통_지출_총금액', '교육_지출_총금액', '유흥_지출_총금액',
                '여가_문화_지출_총금액', '기타_지출_총금액', '음식_지출_총금액']


    for tag, kor_col in zip(tag_list, kor_cols):
        for idx, value in enumerate(response.select(tag)):
            result2.setdefault(kor_col, []).append(value.text)


    print(f"{page_num}/{total_page}수집중입니다", end="\r")
    page_num += 1
    start += 1000
    end += 1000


df = pd.DataFrame(result2)
df
```

```python
df = pd.DataFrame(result2)
df.to_csv("./naver_search_web/data/서울시 상권분석(소득소비-행정동).csv", index=False, encoding="utf-8-sig")
```
