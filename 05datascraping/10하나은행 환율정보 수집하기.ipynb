```python
def new_cols(df):
    new_cols = []
    for col in df.columns:
        if col[0] == col[1] == col[2]:
            new_cols.append(col[0].replace(" ", "_"))
        elif col[0] != col[1]!= col[2]:
            new_cols.append("_".join(col).replace(" ", "_"))
        else:
            new_cols.append("_".join(col[:2]).replace(" ", "_"))
    return new_cols
```
