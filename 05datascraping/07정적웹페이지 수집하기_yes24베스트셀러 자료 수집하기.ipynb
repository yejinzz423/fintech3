# 전체 페이지 수집

```python
import requests
import time
import pandas as pd
from bs4 import BeautifulSoup as bs
```

```python
result = {}
for page in range(1,10):
    
    print(f"{page}/9", end="\r")
    url = "https://www.yes24.com/product/category/bestseller"
    payload = dict(categoryNumber="001", pageNumber=1, pageSize=120)
    r = requests.get(url, params=payload)
    response = bs(r.content, 'lxml')

    book_list = response.select("#yesBestList>li")

    keys = ['제목', '저자', '출판사', '출간일', '가격', '평점', '리뷰수', '상세페이지링크']

    for idx, book in enumerate(book_list):
        book_title = book.select_one(".gd_name").text
        author = text_clean(book.select_one(".authPub.info_auth").text)
        publisher = book.select_one(".authPub.info_pub > a").text
        pub_date = book.select_one(".authPub.info_date").text
        price = int(book.select_one(".yes_b").text.replace(",","")) 
        rating = float(book.select_one(".rating_grade > .yes_b").text) if book.select_one(".rating_grade > .yes_b") != None else 0
        n_reviews = int(book.select_one(".rating_rvCount > a > em.txC_blue").text.replace(",","")) if book.select_one(".rating_rvCount > a > em.txC_blue") != None else 0
        link = "https://www.yes24.com" + book.select_one(".gd_name")['href']
        values = (book_title, author, publisher, pub_date, price, rating, n_reviews, link)

        for key, value in zip(keys, values):
            result.setdefault(key, []).append(value)
            
df = pd.DataFrame(result)
df
```

```python
# 저자 구분
def author_extract(book):

    author = ""
    trans = ""
    article_paint = ""
    paint = ""
    plan = ""
    supervise = ""
    article = ""    
    if "/" not in book.select_one(".authPub.info_auth").text:
        author = (" ".join(book.select_one(".authPub.info_auth").text.strip().split()[:-1]))
    elif "/" in book.select_one(".authPub.info_auth").text:
        items = book.select_one(".authPub.info_auth").text.split("/")
        len_items = len(book.select_one(".authPub.info_auth").text.split("/"))
        if len_items == 2 and "정보 더 보기" in items[0]:
            author = ", ".join(items[1].split()[1:])
        elif len_items >= 2 and "정보 더 보기" not in items[0]:
            author = " ".join(items[0].split()[:-1])
            #print(items[1].split()[-1])
            for item in items[1:]:
                if "역" in item.split()[-1]:
                    trans = " ".join(item.split()[:-1])
                elif "글그림" in item.split()[-1]:
                    article_paint = " ".join(item.split()[:-1])
                elif "그림" in item.split()[-1]:
                    paint = " ".join(item.split()[:-1])
                elif "기획" in item.split()[-1]:
                    plan = " ".join(item.split()[:-1])
                elif "감수" in item.split()[-1]:
                    supervise = " ".join(item.split()[:-1])
                elif "rmf" in item.split()[-1]:
                    article = " ".join(item.split()[:-1])
        
    return author, trans, article_paint, paint,  plan, supervise, article             
```

```python
result = {}
for page in range(1,10):  
    print(f"{page}/9", end="\r")
    url = "https://www.yes24.com/product/category/bestseller"
    payload = dict(categoryNumber="001", pageNumber=1, pageSize=120)
    r = requests.get(url, params=payload)
    response = bs(r.content, 'lxml')
    book_list = response.select("#yesBestList>li")
    keys = ['제목', '저자', '역자', '글그림', '그림', '기획', '감수', '글', 
            '출판사', '출간일', '가격', '평점', '리뷰수', '상세페이지링크']
    for idx, book in enumerate(book_list):
        book_title = book.select_one(".gd_name").text
        author, trans, article_paint, paint, plan, supervise, article = author_extract(book)
        publisher = book.select_one(".authPub.info_pub > a").text
        pub_date = book.select_one(".authPub.info_date").text
        price = int(book.select_one(".yes_b").text.replace(",","")) 
        rating = float(book.select_one(".rating_grade > .yes_b").text) if book.select_one(".rating_grade > .yes_b") != None else 0
        n_reviews = int(book.select_one(".rating_rvCount > a > em.txC_blue").text.replace(",","")) if book.select_one(".rating_rvCount > a > em.txC_blue") != None else 0
        link = "https://www.yes24.com" + book.select_one(".gd_name")['href']
        values = (book_title, author, trans, article_paint, paint, plan, supervise, article, 
                  publisher, pub_date, price, rating, n_reviews, link)
        for key, value in zip(keys, values):
            result.setdefault(key, []).append(value)

df = pd.DataFrame(result)
df
```
