````python
data['class'] = data['class'].apply(lambda x: 1 if x == ' >50K' else 0)
````

````python
cat_cols = data.select_dtypes(include='object')
num_cols = data.select_dtypes(exclude='object')
````

````python
data.columns
````

````python
data.sort_values(by='class')
````

````python
data['class'].unique()
````

````python
data['class']
````

````python
for col in data:
#     print(col)
#     print(data[col].dtype)
    if data[col].dtype == 'object':
        print("="*30, col, "="*30)
        print(data[col].value_counts(), end="\n\n")
        print(data[col].describe(include='all'), end="\n\n")
        data[col].value_counts().plot(kind='bar')
        plt.show()
        # goupby
        cat_goupby = data[[col, 'class']].groupby(col).mean()
        display(cat_goupby.sort_values(by='class', ascending=False))
        print()
    elif data[col].dtype != 'object' and col != 'class':
        print("="*30, col, "="*30)
        print(data[col].describe())
        data[col].plot(kind='hist')
        plt.show()
        data[col].plot(kind='box')
        plt.show()
        # goupby
        num_goupby = data[[col, 'class']].groupby(col).mean()
        display(num_goupby.sort_values(by='class', ascending=False))
        print()
````
