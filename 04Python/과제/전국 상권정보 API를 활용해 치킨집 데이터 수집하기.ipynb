# 전국 치킨집 데이터 정리하기

```python
import json
with open("./data/전국치킨집리스트.json", 'r') as file2:
    chick = file2.read()
data = json.loads(chick)
```

```python
len(data)
```

```python
data[0]['body']['items'][0]
```

```python
data[0]['body']['items'][0].keys()
```

```python
result5 = {}
for page in data:
    for dicts in page['body']['items']:
        for key, value in dicts.items():
            result5.setdefault(key, []).append(value)
print(result5)
```

```python
import pandas as pd
df5 = pd.DataFrame(result5)
df5
```

* 컬럼명을 한국어로 바꿔서 넣기

```python
chick[0]['header']['columns']
```

```python
print(type(chick[1]['header']['columns']))
```

```python
result5 = {}
for page in data:
    for dicts in page['body']['items']:
        for key, value in zip(data[0]['header']['columns'], dicts.values()):
            result5.setdefault(key, []).append(value)
print(result5)
```

```python
import pandas as pd
df5 = pd.DataFrame(result5)
df5
```

# CSV 파이로 저장하기
* 변수명.to_csv("경로와 파일명", index=False, encoding="utf-8-sig")

```python
df5.to_csv("./data/chicken.csv", index=False, encoding="utf-8-sig")
```

# pandas로 JSON 데이터 쉽게 정리하기
* pd.json_normalize(json데이터)

```python
df6 = pd.json_normalize(data[0]['body']['items'])
df6
```

* json_normalize로 만든 데이터프레임을 pages 리스트에 추가

```python
pages = []
for page in data:
    pages.append(pd.json_normalize(page['body']['items']))
```

```python
len(pages)
```

* pandas의 concat 명령어로 여러개의 데이터프레임 하나로 합치기
* pd.concat(데이터프레임 리스트)
* 데이터프레임의 인덱스 초기화 변수.reset_index(drop=True)

```python
df8 = pd.concat(pages)
df8 = df8.reset_index(drop=True)
df8
```
