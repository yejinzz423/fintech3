```python
hotel_info = {}
for index,hotel in enumerate(data):
    print(f"{index}/{len(data)}", end="\r")
    hotel_id = hotel['data']['body']['pdpHeader']['hotelId']
    hotel_name = hotel['data']['body']['propertyDescription']['name']
    hotel_grade = hotel['data']['body']['propertyDescription']['starRating']
    hotel_riviewCount = hotel['data']['body']['reviewContent']['overall']['totalCount']
    hotel_rating = hotel['data']['body']['reviewContent']['overall']['rating']
    cleanliness = hotel['data']['body']['reviewContent']['overall']['ratingAspects'].get('cleanliness', 0.0)
    service = hotel['data']['body']['reviewContent']['overall']['ratingAspects'].get('service', 0.0)
    comfort = hotel['data']['body']['reviewContent']['overall']['ratingAspects'].get('comfort', 0.0)
    condition = hotel['data']['body']['reviewContent']['overall']['ratingAspects'].get('condition', 0.0)
    neighbourhood = hotel['data']['body']['reviewContent']['overall']['ratingAspects'].get('neighbourhood', 0.0)
    hotel_locality = hotel['data']['body']['propertyDescription']['address']['locality']
    hotel_region = hotel['data']['body']['propertyDescription']['address'].get('region', hotel_locality)
   # print(hotel_id,hotel_name,hotel_grade,hotel_riviewCount,hotel_rating,cleanliness,service,comfort,neighbourhood,hotel_region,hotel_locality)
    
    hotel_info.setdefault("hotel_id", []).append(hotel_id)
    hotel_info.setdefault("hotel_name", []).append(hotel_name)
    hotel_info.setdefault("hotel_grade", []).append(hotel_grade)
    hotel_info.setdefault("hotel_riviewCount", []).append(hotel_riviewCount)
    hotel_info.setdefault("hotel_rating", []).append(hotel_rating)
    hotel_info.setdefault("cleanliness", []).append(cleanliness)
    hotel_info.setdefault("service", []).append(service)
    hotel_info.setdefault("comfort", []).append(comfort)
    hotel_info.setdefault("condition", []).append(condition)
    hotel_info.setdefault("neighbourhood", []).append(neighbourhood)
    hotel_info.setdefault("hotel_locality", []).append(hotel_locality)
    hotel_info.setdefault("hotel_region", []).append(hotel_region)

import pandas as pd   
hotel_info_df = pd.DataFrame(hotel_info)
hotel_info_df
```

# 호텔별 사용자 리뷰추출

```python
# 리뷰 위치
reviews = data[0]['data']['body']['reviewContent']['reviews']['hermes']['groups'][0]['items']
```

```python
hotel_reviews = {}
for index,hotel in enumerate(data):
    print(f"{index}/{len(data)}", end="\r")
    try: 
        reviews = hotel['data']['body']['reviewContent']['reviews']['hermes']['groups'][0]['items']
    except Exception as e:
        print(e)
        print("이 호텔은 리뷰가 없습니다.")

    for review in reviews:
        # 호텔 ID
        hotel_id = hotel['data']['body']['pdpHeader']['hotelId']
        # 고객ID
        user_id = review['itineraryId']
        # 리뷰일
        review_date = review['reviewDate']
        # 여행타입
        trip_type = review['tripType']
        # 평점
        rating = review['rating']
        # 리뷰 텍스트
        review_text = review['description']

        # print(hotel_id, user_id, review_date, trip_type, rating, review_text)

        hotel_reviews.setdefault("hotel_id", []).append(hotel_id)
        hotel_reviews.setdefault("user_id", []).append(user_id)
        hotel_reviews.setdefault("review_date", []).append(review_date)
        hotel_reviews.setdefault("trip_type", []).append(trip_type)
        hotel_reviews.setdefault("rating", []).append(rating)
        hotel_reviews.setdefault("review_text", []).append(review_text)
    
```

```python
hotel_reviews_df = pd.DataFrame(hotel_reviews)
hotel_reviews_df
```
